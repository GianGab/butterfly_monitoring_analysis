# Final reducing function
* This function will bring in all the previous functions used for reducing the df for robustness
* Should we need future changes, it should be added to this function


```python
def subset_df(df):
    cleaned_df = df\
    .sample(n=1000000, random_state = 27)\
    .reset_index(drop = True)

    return cleaned_df;

# def subset_df(df, sample_size = 1000000):
#     random_indices = np.random.choice(df.index, size=sample_size, replace=False)
#     sampled_df = df.loc[random_indices]
    
#     return sampled_df

def remove_only_nulls(df):
    ## collect all columns containing only nulls into a list
    full_null_columns = [col for col in df.columns if df[col].isnull().all()]
    cleaned_df = df.drop(axis = 1, columns = df[full_null_columns])
    
    return cleaned_df;


def drop_useless_cols(df):

    useless_columns = ["datasetKey", "kingdom", "phylum", "class", "order", 
                       "taxonRank", "basisOfRecord", "institutionCode", "license", "rightsHolder",
                       "eventDate", "lastInterpreted", "occurrenceStatus", "publishingOrgKey",
                       "issue"]
    cleaned_df = df.drop(columns = df[useless_columns])

    return cleaned_df;


def clean_names_cols(df):
    cleaned_df = df.copy()
    cleaned_df["species"] = [v[1] for v in [x.split() for x in df["species"]]]
    cleaned_df = cleaned_df.drop(columns = "scientificName")\
            .rename(columns = {"verbatimScientificName": "scientificName",
            "verbatimScientificNameAuthorship" : "nameAuthor"})
    
    return cleaned_df;


def reduce_subset(df):
    df_edit_1 = subset_df(df)
    df_edit_2 = remove_only_nulls(df_edit_1)
    df_edit_3 = drop_useless_cols(df_edit_2)
    df_edit_4 = clean_names_cols(df_edit_3)

    return df_edit_4;

```

```python
test_1 = subset_df(df.copy())
print(test_1.shape)
test_2 = remove_only_nulls(test_1)
print(test_2.shape)
test_3 = drop_useless_cols(test_2)
print(test_3.shape)
test_4 = clean_names_cols(test_3)
print(test_4.shape, "\n", test_4.head())
```
